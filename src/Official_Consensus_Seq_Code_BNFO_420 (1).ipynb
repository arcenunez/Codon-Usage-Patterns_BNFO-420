```python


amino_acids = {'ttt': 'F', 'tct': 'S', 'tat': 'Y', 'tgt': 'C',
'ttc': 'F', 'tcc': 'S', 'tac': 'Y', 'tgc': 'C',
'tta': 'L', 'tca': 'S', 'taa': '*', 'tga': '*',
'ttg': 'L', 'tcg': 'S', 'tag': '*', 'tgg': 'W',
'ctt': 'L', 'cct': 'P', 'cat': 'H', 'cgt': 'R',
'ctc': 'L', 'ccc': 'P', 'cac': 'H', 'cgc': 'R',
'cta': 'L', 'cca': 'P', 'caa': 'Q', 'cga': 'R',
'ctg': 'L', 'ccg': 'P', 'cag': 'Q', 'cgg': 'R',
'att': 'I', 'act': 'T', 'aat': 'N', 'agt': 'S',
'atc': 'I', 'acc': 'T', 'aac': 'N', 'agc': 'S',
'ata': 'I', 'aca': 'T', 'aaa': 'K', 'aga': 'R',
'atg': 'M', 'acg': 'T', 'aag': 'K', 'agg': 'R',
'gtt': 'V', 'gct': 'A', 'gat': 'D', 'ggt': 'G',
'gtc': 'V', 'gcc': 'A', 'gac': 'D', 'ggc': 'G',
'gta': 'V', 'gca': 'A', 'gaa': 'E', 'gga': 'G',
'gtg': 'V', 'gcg': 'A', 'gag': 'E', 'ggg': 'G'
}


IUPAC = \
    {
    'A': {'A': 1},
    'C': {'C': 1},
    'G': {'G': 1},
    'T': {'T': 1},
    'R': {'A': 0.5, 'G': 0.5},
    'Y': {'C': 0.5, 'T': 0.5},
    'S': {'G': 0.5, 'C': 0.5},
    'W': {'A': 0.5, 'T': 0.5},
    'K': {'G': 0.5, 'T': 0.5},
    'M': {'A': 0.5, 'C': 0.5},
    'B': {'C': 1/3, 'G': 1/3, 'T': 1/3},
    'D': {'A': 1/3, 'G': 1/3, 'T': 1/3},
    'H': {'A': 1/3, 'C': 1/3, 'T': 1/3},
    'V': {'A': 1/3, 'C': 1/3, 'G': 1/3},
    'N': {'A': 0.25, 'C': 0.25, 'G': 0.25, 'T': 0.25},
    '-': {'-': 1}
    }


def keywithmaxval(d):
    """ a) create a list of the dict's keys and values;
        b) return the key with the max value"""
    v = list(d.values())
    k = list(d.keys())
    return k[v.index(max(v))]

consensus = ''

# seq = []  # These would normally be read in from a file, we can discuss how to do that
# seq.append('ATGGGGCCCAAATTTAGCAGCAGC')
# seq.append('ATGGGGCTCTAATTTAGCAGCAGC')
# seq.append('RTGGGGCCCACATTTAGCAGCATC')

'''
File is of the aligned sequences of either alpha, delta, omic1, or omic2
'''
# Read in file and make seq list
# Make sure file name is correct
with open('omic2_aligned.fasta', 'r') as infile:
    lines = infile.readlines()


# List of the sequences to that will be used to create
# consensus sequence
seq = []
for i in range(1, len(lines), 2):
    x = lines[i].rstrip()
    if x[0] != ">":
        seq.append(x)
print(len(seq))
# Go through each sequence
for pos in range(len(seq[0])):

    symbol_dict = {}
    # Go through each position of the sequence
    for cur_seq in seq:

        symbol = cur_seq[pos]
        # print('Cur symbol', symbol)
        # print(IUPAC[symbol].items())
        for sub_symbol, count in IUPAC[symbol].items():
            # print(sub_symbol, count), '/n'

            try:
                symbol_dict[sub_symbol] += count
            except KeyError:
                symbol_dict[sub_symbol] = count

    # print(symbol_dict)
    consensus += keywithmaxval(symbol_dict)
# Check if len of consensus is the same as the ref
print(len(consensus))
# Print consensus (if you want to look at it)
print(consensus)
'''
New file name must match whichever variant we are creating the
consensus seq for
'''
# Write consensus sequence to a file
# RENAME file for approriate consensus sequence
with open("omic2_consensus.fasta", "w") as file:
    file.write(consensus)
```
